## Abstract

This report discusses two methods to implement more precise multiplications with the NEF. First, it will be shown that encoders aligned to the diagonals produce better results than the usual randomly distributed encoders. Second, the distribution of evaluation points will be considered. A diamond shape is shown to be best.

```python
import nengo
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

%matplotlib inline
```

## A benchmark

To compare different methods of multiplication we will first define a benchmark used throughout this report. We assume that the product of two independent scalars in the range $[-1, 1]$ has to be calculated. This is a reasonable default assumption as $[-1, 1]$ is the default representational range used in the NEF and without additional information there is no reason to assume any correlation of the input values. If, however, the distribution of input violates this assumption (e.g. they fall into a unit circle instead of a unit square), the best distribution of evaluation points given below might differ.

For the benchmark the whole range from $[-1, 1]$ should be covered for both values and number of combinations of different values should occur. Both values should contious over time because dicontinueities will introduce errors not related to the calculation of the product itself. Using a linear function and a sine wave for the multiplicands is a function satisfying these requirements.

```python
def stimulus_fn(t):
    x = 2 * t - 1
    y = np.sin(np.pi * x * 5)
    return np.array([x, y])
```

This is what the input looks like:

```python
plt.subplot(1, 1, 1, aspect='equal')
plt.plot(*stimulus_fn(np.linspace(0, 1, 1000)))
plt.xlabel("x")
plt.ylabel("y");
```

We use this input to feed it into the product ensemble under test. From the decoded output and an ensemble calculating the same product in direct mode the root mean square error (RMSER) is computed as a performance indicator.

```python
from nengo.utils.numpy import rmse

class Benchmark(object):
    def __init__(self, seed, n_neurons=50, **ens_conf):
        self.n_neurons = n_neurons
        self.seed = seed

        multiply = lambda x: x[0] * x[1]

        model = nengo.Network(seed=seed)
        with model:
            # Note that we set the radius to sqrt(2.) to make sure all possible input values
            # are within the representational range of the ensemble. (Both multiplicands
            # could be 1, giving a distance of sqrt(1²+1²) to the origin.)
            ens_lif = nengo.Ensemble(n_neurons, dimensions=2, radius=np.sqrt(2.), **ens_conf)
            self.e = ens_lif
            ens_direct = nengo.Ensemble(1, dimensions=2, neuron_type=nengo.Direct())

            stimulus = nengo.Node(output=stimulus_fn, size_out=2)
            nengo.Connection(stimulus, ens_lif)
            nengo.Connection(stimulus, ens_direct)

            result_lif = nengo.Node(size_in=1)
            nengo.Connection(ens_lif, result_lif, function=multiply)

            result_direct = nengo.Node(size_in=1)
            nengo.Connection(ens_direct, result_direct, function=multiply)

            self.probe_lif = nengo.Probe(result_lif)
            self.probe_direct = nengo.Probe(result_direct)
            self.p = nengo.Probe(ens_lif)
            
        self.sim = nengo.Simulator(model)
        self.sim.run(1.)
        
    def rmse(self):
        return rmse(self.sim.data[self.probe_lif], self.sim.data[self.probe_direct])
        
    def plot(self, ax=None):
        if ax is None:
            ax = plt.gca()
        ax.plot(self.sim.trange(), self.sim.data[self.probe_lif], label="LIF")
        ax.plot(self.sim.trange(), self.sim.data[self.probe_direct], label="Direct")
        ax.legend(loc='best')
        ax.set_xlabel("t [s]")
```

```python
seeds = [3287, 32, 35626, 78901, 684]

def repeated_benchmark(**ens_conf):
    benchmarks = [Benchmark(s, **ens_conf) for s in seeds]
    benchmarks[0].plot()
    rmse = np.mean([b.rmse() for b in benchmarks])
    print "RMSE:", rmse
    return rmse
```

### The Naive implementation

The naive implementation with the default configuration of an ensemble gives us a baseline. Note that a radius of $\sqrt{2}$ is used to ensure all possible input values fall into the representational range of the ensemble.

```python
rmse_naive = repeated_benchmark()
```

## Encoders

The largest reduction in error is obtained by using diagonal encoders. To see why this is the case we analytically derive the error in different directions of the input space.

Assume the the multiplicands are given as a vector $\vec x = (x_1, x_2)^{\top}$. Then the result of the product can be rewritten in polar coordinates:

$$x_1 x_2 = r^2 \cos \varphi \sin \varphi$$

Furthermore, it is neccessary to assume a special neuron model in the derivations. For every input on its encoder $\vec e = (e_1, e_2)^{\top}, |\vec e| = 1$ it calculates a perfect product, but the input current proportional to $\vec e \cdot \vec x$ is rectified:

$$ a(\vec x) = e_1 [\vec e \cdot \vec x]_+ e_2 [\vec e \cdot \vec x]_+ $$

With $\psi$ as the polar coordinate of the encoder $\vec e$ this can be rewritten as

$$ a(\vec x) = r^2 \cos \psi \sin \psi [\cos(\psi - \varphi)]_+^2 $$

This allows to write down the average error for multiplications with multiplicands aligned along $\varphi$.

$$ E^2(\varphi) = \langle (x_1 x_2 - \sum_i a_i(\vec x))^2 \rangle_r \\
 = \langle (r^2 \cos \varphi \sin \varphi - r^2 \sum_i \cos \psi_i \sin \psi_i [\cos(\psi_i - \varphi)]_+^2)^2 \rangle_r \\
 = \frac{R^3}{4} (\cos \varphi \sin \varphi - \sum_i \cos \psi_i \sin \psi_i [\cos(\psi_i - \varphi)]_+^2)^2$$
 
In this equation $R$ is the maximum radius of the NEF ensemble used for the multiplication. Note that it is only a constant factor not altering the qualitative distribution of the error.

```python
def mult_err(phi, encoders=[], R=np.sqrt(2.)):
    if len(encoders) > 0:
        cos_sin_psi = np.cos(encoders) * np.sin(encoders)
        diff = np.maximum(0., np.cos(np.subtract.outer(encoders, phi)))
        neuron_outputs = cos_sin_psi[:, np.newaxis] * diff ** 2.
    else:
        neuron_outputs = np.zeros((0, np.asarray(phi).size))
    return R ** 3. / 4. * (np.cos(phi) * np.sin(phi) - np.sum(neuron_outputs, axis=0))  ** 2.
```

Without any neurons we get the following distribution of the error:

```python
phis = np.linspace(0., 2. * np.pi, 100)
ax = plt.subplot(1, 1, 1, polar=True)
ax.plot(phis, mult_err(phis))
ax.set_rmax(0.2)
ax.yaxis.set_major_locator(MaxNLocator(3))
```

As we see the maxima of the error are aligned with the diagonals. By adding neurons with each diagonal as an encoder one after another the error gets reduced to zero:

```python
encoders = [np.pi / 4., 3. / 4. * np.pi, 5. / 4. * np.pi, 7. / 4. * np.pi]

plt.figure(figsize=(20, 8))
for i in range(1, len(encoders) + 1):
    ax = plt.subplot(1, len(encoders) + 1, i + 1, polar=True)
    ax.plot(phis, mult_err(phis, encoders=encoders[:i]))
    ax.set_rmax(0.2)
    ax.yaxis.set_major_locator(MaxNLocator(3))
    ax.set_title("{0} encoders".format(i))
```

Using encoders with a different alignment will give worse results as the maximal possible reduction in error is less than for the diagonal directions. Using those diagonal encoders instead fo random encoders in the benchmark we obtain a substantial reduction of the error.

```python
diag_encoders = nengo.dists.Choice(np.array([[1, 1], [1, -1], [-1, -1], [-1, 1]]) / np.sqrt(2.))
rmse_diag_encoders = repeated_benchmark(encoders=diag_encoders)
print "That is an improvement by {0:.0f}%.".format((1. - rmse_diag_encoders / rmse_naive) * 100.)
```

## Evaluation points

The evaluation points are sampled from a circle with radius $r$ per default. As we assumed independent inputs we set $r = \sqrt{2}$ to have values like $\vec x = (1, 1)^{\top}$ in range. This, however, will produce a number of evaluation points outside the allowed input range. Here is a visualization with the range of possible inputs as the red square and the area from which the evaluation points are chosen as blue cirle.

```python
plt.subplot(1, 1, 1, aspect='equal')
plt.gca().add_artist(plt.Circle([0, 0], radius=np.sqrt(2), fill=False, color='b'))
plt.gca().add_artist(plt.Rectangle([-1, -1], 2, 2, fill=False, color='r'))
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5);
```

It seems that we can achieve a more accurate product by sampling the evaluation points from the square of actual possible inputs instead of “wasting” some in an irrelevant area.

```python
rmse_square = repeated_benchmark(eval_points=nengo.dists.Uniform(-1 / np.sqrt(2.), 1 / np.sqrt(2.)))
print "That is an improvement by {0:.0f}%.".format((1. - rmse_square / rmse_naive) * 100.)
```

Suprisingly, it does not make a huge difference and seems to be even a little bit worse. But maybe it helps together with the diagonal encoders?

```python
rmse_square_diag_encoders = repeated_benchmark(
    encoders=diag_encoders, eval_points=nengo.dists.Uniform(-1 / np.sqrt(2.), 1 / np.sqrt(2.)))
print "That is an improvement by {0:.0f}%.".format((1. - rmse_square_diag_encoders / rmse_diag_encoders) * 100.)
```

The opposite is the case; the error doubled! To understand why this is the case it is necessary to realize that the evaluation points are projected onto the encoders. The following code block samples evaluation points from a circle and a square and visualizes the distribution of the values projected onto the first diagonal encoder.

```python
n = 20000
d = 2
circle = nengo.dists.UniformHypersphere().sample(n, d) * np.sqrt(2.)
square = nengo.dists.Uniform(-1, 1).sample(n, d)

encoder = np.ones(d) / np.sqrt(2.)
bins = np.linspace(-np.sqrt(2.), np.sqrt(2.), 50)

plt.hist(np.dot(circle, encoder), bins=bins, alpha=0.5, normed=True, label='circle')
plt.hist(np.dot(square, encoder), bins=bins, alpha=0.5, normed=True, label='square')
plt.legend();
```

The distribution obtained from the circle is broader. That means a greater proportion of values have large absolute projected values. Inputs projecting to large absolute values can contribute to the multiplication error more than smaller values because the multiplication result changes more quickly for large multiplicands. Thus, even though the values are outside the input range when sampling from a circle, these values can improve the function approximation.

So far, this description was informal. Next we will analytically derive the expected error given a point in the input space. Let the inputs be given as $\vec x = (x_1, x_2)^{\top}$. The true result of the multiplication is than $x_1 x_2$. However, there will be noise from different sources in the system:

1. The representation is noisy and $\vec x$ will not be perfectly represented.
2. The number of evaluation points is finite. The decoded function between those points can be off (how much is limited by the number and shape of the neuron's tuning curves).
3. The degrees of freedom for approximating the decoded function is limited by the number of neuron's and their tuning curves. Thus, the true function can not be matched perfectly in general.

For the following derivation additive normal distributed noise which is independent for both components is assumed, i.e. $(x_1 + \eta_1, x_2 + \eta_2)^{\top}$ with $\eta_1, \eta_2 \sim \mathcal{N}(0, \sigma^2)$. This allows to derive the squared expected error $E^2(\vec x)$ as

$$E^2(\vec x) = \mathbb{E}[(x_1 x_2 - (x_1 + \eta_1)(x_2 + \eta_2))^2] \\
= \mathbb{E}[x_1^2 x_2^2 - 2 x_1 x_2 (x_1 + \eta_1)(x_2 + \eta_2) + (x_1 x_2 + x_1 \eta_2 + x_2 \eta_1 + \eta_1 \eta_2)^2] \\
= x_1^2 \mathbb{E}[\eta_2^2] + x_2^2 \mathbb{E}[\eta_1^2] + \mathbb{E}[\eta_1^2] \mathbb{E}[\eta_2^2] \\
= (x_1^2 + x_2^2) \sigma^2 + \sigma^4$$

Furthermore, this error function can be normalized to a probability distribution function (PDF):

$$p_{E^2}(\vec x) = \frac{E^2(\vec x)}{\frac{8}{3} \sigma^2 + 4 \sigma^4} = \frac{x_1^2 + x_2^2 + \sigma^2}{\frac{8}{3} + 4 \sigma^2}$$

```python
def sqerr_pdf(x1, x2, sigma=0.01):
    return (x1 ** 2. + x2 ** 2. + sigma ** 2.) / (8. / 3. + 4. * sigma ** 2.)
```

By visualizing the PDF it becomes evident that the largest errors occur in the corners of the square. (An alternative interpretation would be that the probability for an error is highest in those areas.) The projection onto the diagonal encoders yields large magnitudes for the corners. This is in line with the previous observations.

```python
spacing = np.linspace(-1, 1, 50)
x1, x2 = np.meshgrid(spacing, spacing, indexing='ij')
plt.title(r"$p_{E^2}(\vec x)$")
plt.imshow(sqerr_pdf(x1, x2), vmin=0., extent=[-1, 1, -1, 1], cmap='Blues')
plt.colorbar();
```

The reason for normalizing to a PDF is that it allows us to easily sample from the distribution with rejection sampling.

```python
class SqErrDist(nengo.dists.Distribution):
    def __init__(self, sigma=0.01):
        self.sigma = sigma
        
    def sample(self, n, d, rng=np.random):
        assert d == 2
        x = np.empty((n, d))
        for i in range(n):
            p = 0.
            while p < rng.rand():
                x[i, :] = nengo.dists.Uniform(-1, 1).sample(1, d, rng)
                p = sqerr_pdf(x[i, 0], x[i, 1], sigma=self.sigma)
        return x /  np.sqrt(2.)
```

```python
x = SqErrDist().sample(1000, 2) * np.sqrt(2.)
plt.subplot(1, 1, 1, aspect='equal')
plt.title(r"Samples from $p_{E^2}(\vec x)$")
plt.scatter(x[:, 0], x[:, 1], marker='+')
plt.xlim(-1, 1)
plt.ylim(-1, 1);
```

These evaluation point actually give an improvement. Though, it is small (about 3%) when diagonal encoders are used.

```python
rmse_sqerr_dist = repeated_benchmark(eval_points=SqErrDist())
print "That is an improvement by {0:.0f}%.".format((1. - rmse_sqerr_dist / rmse_naive) * 100.)
```

```python
rmse_sqerr_dist_diag_encoders = repeated_benchmark(eval_points=SqErrDist(7.5), encoders=diag_encoders)
print "That is an improvement by {0:.0f}%.".format((1. - rmse_sqerr_dist_diag_encoders / rmse_square_diag_encoders) * 100.)
```

The careful reader will have noticed that there is a free parameter in the error function: the noise variance $\sigma^2$. Right now there is no way of knowing the true value of $\sigma^2$. Above we used $\sigma = 0.01$ which seems to work well. Larger values will shift the distribution of evaluation points towards a uniform distribution eliminating the improvements. Smaller values shift the distribution away from the uniform distribution. Given the normalization to a PDF it is possible to see what happens for $\sigma \rightarrow 0$.

$$ \lim_{\sigma \rightarrow 0} p_{E^2}(\vec x) = \frac{3}{8} (x_1^2 + x_2^2)$$

In fact, the result obtained with $\sigma = 0$ is almost as good as with $\sigma = 0.05$ (without diagonal encoders even better):

```python
class UniEnc(object):
    def sample(self, n, d):
        samples = np.empty((n, d))
        for i in range(n):
            a = 0.
            b = 1.
            while a < 0.5 and np.abs(b) > a:
                a = np.random.uniform(0., 1.)
                b = np.random.uniform(a - 1., 1. - a)
            x = np.random.uniform(0., 1.)
            if x < 0.25:
                a_basis = np.array([1., 1.])
                b_basis = np.array([-1., 1.])
            elif x < 0.5:
                a_basis = np.array([-1., -1.])
                b_basis = np.array([-1., 1.])
            elif x < 0.75:
                a_basis = np.array([-1., 1.])
                b_basis = np.array([1., 1.])
            else:
                a_basis = np.array([1., -1.])
                b_basis = np.array([1., 1.])
            samples[i] = a * a_basis + b * b_basis
        return samples
```
