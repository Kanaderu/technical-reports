## Abstract

This report discusses two methods to implement more precise multiplications with the NEF. First, it will be shown that encoders aligned to the diagonals produce better results than the usual randomly distributed encoders. Second, the distribution of evaluation points will be considered. A diamond shape is shown to be best.

```python
import nengo
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## A benchmark

To compare different methods of multiplication we will first define a benchmark used throughout this report. We assume that the product of two independent scalars in the range $[-1, 1]$ has to be calculated. This is a reasonable default assumption as $[-1, 1]$ is the default representational range used in the NEF and without additional information there is no reason to assume any correlation of the input values. If, however, the distribution of input violates this assumption (e.g. they fall into a unit circle instead of a unit square), the best distribution of evaluation points given below might differ.

For the benchmark the whole range from $[-1, 1]$ should be covered for both values and number of combinations of different values should occur. Both values should contious over time because dicontinueities will introduce errors not related to the calculation of the product itself. Using a linear function and a sine wave for the multiplicands is a function satisfying these requirements.

```python
def stimulus_fn(t):
    x = 2 * t - 1
    y = np.sin(np.pi * x * 5)
    return np.array([x, y])
```

This is what the input looks like:

```python
plt.subplot(1, 1, 1, aspect='equal')
plt.plot(*stimulus_fn(np.linspace(0, 1, 1000)))
plt.xlabel("x")
plt.ylabel("y")
```

We use this input to feed it into the product ensemble under test. From the decoded output and an ensemble calculating the same product in direct mode the root mean square error (RMSER) is computed as a performance indicator.

```python
from nengo.utils.numpy import rmse

class Benchmark(object):
    def __init__(self, seed, n_neurons=50, **ens_conf):
        self.n_neurons = n_neurons
        self.seed = seed

        multiply = lambda x: x[0] * x[1]

        model = nengo.Network(seed=seed)
        with model:
            # Note that we set the radius to sqrt(2.) to make sure all possible input values
            # are within the representational range of the ensemble. (Both multiplicands
            # could be 1, giving a distance of sqrt(1²+1²) to the origin.)
            ens_lif = nengo.Ensemble(n_neurons, dimensions=2, radius=np.sqrt(2.), **ens_conf)
            ens_direct = nengo.Ensemble(1, dimensions=2, neuron_type=nengo.Direct())

            stimulus = nengo.Node(output=stimulus_fn, size_out=2)
            nengo.Connection(stimulus, ens_lif)
            nengo.Connection(stimulus, ens_direct)

            result_lif = nengo.Node(size_in=1)
            nengo.Connection(ens_lif, result_lif, function=multiply)

            result_direct = nengo.Node(size_in=1)
            nengo.Connection(ens_direct, result_direct, function=multiply)

            self.probe_lif = nengo.Probe(result_lif)
            self.probe_direct = nengo.Probe(result_direct)
            
        self.sim = nengo.Simulator(model)
        self.sim.run(1.)
        
    def rmse(self):
        return rmse(self.sim.data[self.probe_lif], self.sim.data[self.probe_direct])
    
    def plot(self, ax=None):
        if ax is None:
            ax = plt.gca()
        ax.plot(self.sim.trange(), self.sim.data[self.probe_lif], label="LIF")
        ax.plot(self.sim.trange(), self.sim.data[self.probe_direct], label="Direct")
        ax.legend(loc='best')
        ax.set_xlabel("t [s]")
```

```python
seeds = [3287, 32, 35626, 78901, 684]

def repeated_benchmark(**ens_conf):
    benchmarks = [Benchmark(s, **ens_conf) for s in seeds]
    benchmarks[0].plot()
    print "RMSE:", np.mean([b.rmse() for b in benchmarks])
```

### The Naive implementation

The naive implementation with the default configuration of an ensemble gives us a baseline. Note that a radius of $\sqrt{2}$ is used to ensure all possible input values fall into the representational range of the ensemble.

```python
repeated_benchmark()
```

## Encoders

The largest reduction in error is obtained by using diagonal encoders. To see why this is the case we analytically derive the error in different directions of the input space.

Assume the the multiplicands are given as a vector $\vec x = (x_1, x_2)^{\top}$. Then the result of the product can be rewritten in polar coordinates:

$$x_1 x_2 = r^2 \cos \varphi \sin \varphi$$

Furthermore, it is neccessary to assume a special neuron model in the derivations. For every input on its encoder $\vec e = (e_1, e_2)^{\top}, |\vec e| = 1$ it calculates a perfect, but rectified product:

$$ a(\vec x) = [e_1 (\vec e \cdot \vec x) e_2 (\vec e \cdot \vec x)]_+ $$

With $\psi$ as the polar coordinate of the encoder $\vec e$ this can be rewritten as

$$ a(\vec x) = r^2 [\cos \psi \sin \psi \cos^2(\psi - \varphi)]_+ $$

This allows to write down the error for a given $\varphi$.

$$ E^2(\varphi) = \langle (x_1 x_2 - \sum_i a_i(\vec x))^2 \rangle_r \\
 = \langle (r^2 \cos \varphi \sin \varphi - r^2 \sum_i [\cos \psi_i \sin \psi_i \cos^2(\psi_i - \varphi)]_+)^2 \rangle_r \\
 = \frac{R^3}{4} (\cos \varphi \sin \varphi - \sum_i [\cos \psi_i \sin \psi_i \cos^2(\psi_i - \varphi)]_+)^2$$

```python
repeated_benchmark(encoders=nengo.dists.Choice([[1, 1], [1, -1], [-1, -1], [-1, 1]]))
```

## Evaluation points

```python
eval_points = nengo.dists.Uniform(-np.sqrt(2.), np.sqrt(2.)).sample(1000, 2)
eval_points = np.dot(np.array(
    [[np.cos(np.pi/4), -np.sin(np.pi/4)],
     [np.sin(np.pi/4), np.cos(np.pi/4)]]), eval_points.T).T
plt.scatter(eval_points[:, 0], eval_points[:, 1])
```

```python
repeated_benchmark(eval_points=eval_points)
```

```python
repeated_benchmark(eval_points=eval_points, encoders=nengo.dists.Choice([[1, 1], [1, -1], [-1, -1], [-1, 1]]))
```
